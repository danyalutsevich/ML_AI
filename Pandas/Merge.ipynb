```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({"employee":["Bob","Jake","Lisa","Sue"],
"group":["Accounting","Engeneering","Engeneering","HR"]
})

df2 = pd.DataFrame({"employee":["Lisa","Bob","Jake","Sue"],
"hire_date":[2004,2008,2012,2014]
})
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({
    "group":["Accounting","Engeneering","HR"],
    "supervisor":["Carly","Guido","Steve"]
})
df4
```

```python
pd.merge(df3,df4)
```

```python
df5 = pd.DataFrame({
    "group":["Accounting","Accounting","Engeneering","Engeneering","HR","HR"],
    "skills":["math","spreadsheets","coding","linux","spreadsheets","organization"]
})
df5
```

```python
pd.merge(df1,df5)
```

```python
df2_1 = pd.DataFrame({"employee":["Lisa","Bob","Jake","Sue"],
"group":[2004,2008,2012,2014],
"hire_date":[2004,2008,2012,2014],
})
df3_1 = pd.merge(df1,df2_1,on="employee")
df3_1
```

```python
df2_2 = pd.DataFrame({"employee":["Jake","Sue"],
"hire_date":[2004,2008],
})
```

```python
pd.merge(df1,df2_2,on="employee")
```
