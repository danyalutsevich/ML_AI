```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("../data/titanic.csv",delimiter=";")
```

```python
titanic.head()
```

```python
titanic.groupby(["Sex","Class"])["Survived"].mean().unstack()
```

```python
titanic.pivot_table(values="Survived",index="Sex",columns="Class")
```

```python
titanic.pivot_table(index="Sex")
```

```python
titanic.pivot_table(index=["Sex","Class"],values="Freq")
```

```python
titanic.pivot_table(index=["Sex","Class"],values="Freq",aggfunc=max)
```

```python
titanic.pivot_table(index="Sex",columns="Class",values="Survived",aggfunc=[len,lambda x:x.sum()/x.count()*100])
```

```python
def percent_not_null(val:pd.Series):
    n=val.notnull().count()
    m=val.count()
    return n/m*100
```

```python
titanic.pivot_table(index="Sex",columns="Class",values="Survived",aggfunc=[len,percent_not_null],margins=True,margins_name="Total")
```

```python
titanic.pivot_table(index="Sex",columns="Class",values=["Survived","Freq"],aggfunc=[len,np.mean])
```
