```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"key": ["A", "B", "C", "A", "B", "C"],
                   "data": range(6)}, columns=["key", "data"])
df
```

```python
df.groupby("key").sum()
```

```python
df["new"] = np.random.randint(100, 350, size=[6, 1])
df
```

```python
df.groupby("key").median()
```

```python
df.groupby("key")["new"].median()
```

```python
df.groupby("key")["new"].describe()
```

```python
df.groupby("key").aggregate(["min", np.median, max])
```

```python
df.groupby("key")["new"].aggregate([np.mean, np.std])
```

```python
df.groupby("key")["new"].aggregate(
    [min, np.max, lambda x:np.percentile(x, 0.75)])
```

```python
df.groupby("key").aggregate({"data": [np.mean, np.std], "new": [
    min, np.max, lambda x:np.percentile(x, 0.75)]})
```

```python
def filter_func(x):
    return x["new"].std() > 10
```

```python
df.groupby('key').filter(filter_func)
```

```python
df.groupby("key").filter(lambda x: x["data"].mean() > 2)
```

```python
df1 = pd.DataFrame({
    "type": ["book", "game", "food", "game", "game", "food", "book"],
    "name": ["item1", "item2", "item3", "item4", "item5", "item6", "item7"],
    "price": [120, 87, 100, 34, 210, 1000, 56],
    "amount": [2, 3, 23, 50, 3, 5, 100]
})

df1
```

```python
df1.groupby("type").filter(lambda x: x["price"].max() > 200)
```

```python
df1.groupby("type").filter(
    lambda x: x["amount"].mean() > 5 and x["amount"].mean() < 50)
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
            'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
            'Points': [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}
df2 = pd.DataFrame(ipl_data)
df2
```

```python
df2.groupby("Team").filter(lambda x: len(x) >= 3)
```

```python
df2.groupby("Team").transform(lambda x: x-x.mean())
```

```python
df3 = pd.DataFrame({
    "type": ["book", "game", "food", "game", "game", "food", "book", "game", "food", "book"],
    "name": ["item1", "item2", "item3", "item4", "item5", "item6", "item7", "item8", "item9", "item10"],
    "price": [120, 87, 100, 34, 210, 1000, 56, 4, 23, 68],
    "amount": [2, 3, 23, 50, 3, 5, 100, 33, 2, 78]
})

df3
```

```python
foodDict = {"game": "notfood", "book": "notfood", "food": "food" }
df3 = df3.set_index("type")
```
