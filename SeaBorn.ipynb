```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
pokemons = pd.read_csv("../data/pokemon1.csv",index_col=0,encoding_errors='ignore')
```

```python
pokemons.info()
```

```python
pokemons
```

```python
sns.lmplot(x="Attack",y="Defense",data=pokemons)
```

```python
sns.lmplot(x="Attack",y="Defense",data=pokemons,fit_reg=False,hue="Stage")
plt.ylim(0,None)
plt.xlim(0,None)
```

```python
sns.boxplot(data=pokemons)
```

```python
stats = pokemons.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=stats)
```

```python
sns.boxplot(x="Stage",y="Attack",data=pokemons,palette="coolwarm")
```

```python
sns.boxplot(x="Stage",y="Attack",hue="Legendary",data=pokemons,palette="coolwarm")
```

```python
sns.violinplot(x="Type 1",y="Attack",data=pokemons)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x="Type 1",y="Attack",data=pokemons,palette=pkmn_type_colors)
```

```python
sns.swarmplot(x="Type 1",y="Attack",data=pokemons,palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x="Type 1",y="Attack",data=pokemons,palette=pkmn_type_colors,inner=None)

sns.swarmplot(x="Type 1",y="Attack",data=pokemons,color='k',alpha=0.7)
plt.title('Attack by Type')
```

```python
corr = pokemons.corr()
sns.heatmap(corr)
```
