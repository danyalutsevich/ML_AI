```python
import requests
import os
import re
from collections import Counter
import stop_words
import numpy as np
```

```python
stop = stop_words.get_stop_words('en')
path = './data/Pulp_Fiction.txt'
replace_symbols = [',','.','"',"'",'1','2','3','4','5','6','7','8','9','0',')','(','-',':',';','/','<','>','?','!']
```

```python
def get_words_bag(path, replace):
    text = open(path, "r").read().lower()
    for r in replace:
        text = text.replace(r, '')
    words = text.split()
    bag = []
    for w in words:
      if w not in stop:
         bag.append(w)
    return bag
```

```python
words = get_words_bag(path,replace_symbols)
```
