```python
import requests
import os
import re
from collections import Counter
import stop_words
import numpy as np
```

```python
stop = stop_words.get_stop_words('en')
paths = [
         '../data/scripts/Pulp_Fiction.txt',
         '../data/scripts/Inglorious_Basterds.txt',
         '../data/scripts/Django_Unchained.txt',
         '../data/scripts/Kill_Bill_Vol_1_and_2.txt',
         '../data/scripts/Reservioir_Dogs.txt',
         '../data/scripts/The_Godfather.txt',
         '../data/scripts/The_Godfather_2.txt',
         ]
replace_symbols = [',','.','"',"'",'1','2','3','4','5','6','7','8','9','0',')','(','-',':',';','/','<','>','?','!',']','[','{','}','|','@','#','$','%','^','&','*','_','+','=','~','`','\\','\t','\r','\n',"“"]
```

```python
def get_words_bag(path, replace):
    text = open(path, "r").read().lower()
    for r in replace:
        text = text.replace(r, '')
    words = text.split()
    bag = []
    for w in words:
      if w not in stop:
         bag.append(w)
    return bag
```

```python
texts = []
for p in paths:
    texts.append(get_words_bag(p, replace_symbols))
texts
```

```python
# word | count | appearances in other documents
```

```python
document_size = []
for w in texts:
    document_size.append(len(w))
print(document_size)
print(np.sum(document_size))
```

```python
def get_all_words(texts):
    all_words = []
    for i in range(len(texts)):
        for w in texts[i]:
            all_words.append(w)
    return all_words
all_words = get_all_words(texts)
all_words
```

```python
all_words_with_frequencies = np.array(Counter(all_words).most_common())
all_words_with_frequencies
all_words_with_frequencies[:,0]
```

```python
frequencies_in_texts = []
for text in texts:
    frequencies_in_texts.append(np.array(Counter(text).most_common()))
frequencies_in_texts
```

```python
texts_data = []
for text_with_freqs in frequencies_in_texts:
    text_data=[]
    for word in text_with_freqs:
        text_data.append([word[0],word[1],all_words_with_frequencies[np.where(all_words_with_frequencies[:,0]==word[0])][0][1]])
    texts_data.append(text_data)
```
