```python
import numpy as np
```

```python
LR = 0.3
EPOCHS = 1000
B = 5
Bn = np.random.randint(low=-100, high=100)
```

```python
def neural_network(Bn):
    return Bn
```

```python
def K(model, expected):
    return 2 * (model - expected)
```

```python
def b_predictor(Bn):
    interations = 0
    while interations < EPOCHS:
        interations += 1
        k = K(neural_network(Bn), B)
        Bn = Bn - LR * k
        print(f"{interations} B = ",Bn,f"accuracy {int((Bn*100)/B)}%",)    
```

```python
b_predictor(Bn)
```
