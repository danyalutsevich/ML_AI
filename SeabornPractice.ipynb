```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips)
```

```python
sns.lmplot(x='size', y='tip',data=tips)
```

```python
sns.lmplot(x='size', y='tip',data=tips, fit_reg=False, hue="smoker")
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips, fit_reg=False, hue="smoker")
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

```python
sns.boxplot(x='time', y='tip', data=tips)
```

```python
sns.boxplot(x='size', y='tip', data=tips,palette="coolwarm")
```

```python
sns.boxplot(x='sex', y='tip', data=tips,palette="coolwarm", hue="smoker")
```

```python
sns.violinplot(x='sex', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='sex', y='total_bill', data=tips)
```
