```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
# %matplotlib inline
```

```python
real_w = 2
real_b = 1
LR = 0.1
N = 100
np.random.seed(33)

x = np.random.rand(N,1)

epsilon  = np.random.randn(N,1) * 0.1

y = real_w * x + real_b + epsilon
plt.scatter(x,y)
```

```python
indices = np.arange(N)
np.random.shuffle(indices)
train_indices = indices[:int(N*0.8)]
value_indices = indices[int(N*0.8):]


x_train = x[train_indices]
y_train = y[train_indices]

x_value = x[value_indices]
y_value = y[value_indices]

plt.scatter(x_train,y_train)
plt.scatter(x_value,y_value)
```

```python
w = np.random.randn(1)
b = np.random.randn(1)

print(w,b)

y_model = w * x_train + b

plt.scatter(x_train,y_train)
plt.scatter(x_train,y_model)
```

```python

error = y_model - y_train
loss = (error ** 2).mean()
print(loss)

w_grad = 2 * (error * x_train).mean()
b_grad = 2 * error.mean()


w = w - LR * w_grad
b = b - LR * b_grad

y_model = w * x_train + b

print("w", w, "b", b)

plt.scatter(x_train, y_train)
plt.scatter(x_train, y_model)
```
